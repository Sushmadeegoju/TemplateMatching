**Data inclusion**

```python
data=pd.read_csv("../input/A_Z Handwritten Data.csv")
temp=np.asarray(data)
```

**Matching function**

```python
def matching_func(input_arr,template_arr):
    temp=0
    temp=np.sum((input_arr-template_arr)*(input_arr-template_arr))
    return temp
    
```

**Splitting into train, validation and test sets**

```python
count=[0]*26
for i in range(len(temp)):
    count[temp[i][0]]+=1
train=[]
validate=[]
test=[]
start=0
for i in range(len(count)):
    train1=round(0.6*count[i])
    val1=test1=round(0.2*count[i])
    end=start+train1+val1+test1
    if(end<=len(temp)+2):
        for j in range(start,end):
            if(j<(start+train1)):
                train.append(temp[j][:])
        
            elif(j<(start+train1+val1)):
                validate.append(temp[j][:])
               
            elif(j<(start+train1+val1+test1)):
                if(j<372450):
                    test.append(temp[j][:])
                
        start=end
```

**Average templates**

```python
avg=np.zeros((26,785))
character=train[0][0]
for i in range(len(train)):
    if(train[i][0]==character):
        avg[character]+=train[i]
    else:
        avg[character]/=(count[character]*0.6)
        character=train[i][0]
        i-=1
    
```

**Average bitmap**

```python
for i in range(len(avg)):
    avg[i][0]=round(avg[i][0])
    for j in range(1,785):
        if (avg[i][j]>=128):
            avg[i][j]=1
        else:
            avg[i][j]=0
```

**Test Bitmap**

```python
for i in range(len(test)):
        for j in range(1,785):
            if (test[i][j]>=128):
                test[i][j]=1
            else:
                test[i][j]=0
```

**Prediction**

```python
x=len(test)
predict=[0]*x
for i in range(len(test)):
    min_match=math.inf
    for j in range(len(avg)):
        temp=matching_func(test[i],avg[j])
        if(temp<=min_match):
            min_match=temp
            index=j
    predict[i]=avg[index][0]
```

**Accuracy**

```python
count8=0
for i in range(len(predict)):
    if(test[i][0]==predict[i]):
        count8+=1
print((count8/len(test))*100)
```
